Task 2

The cost column in the dataframe “Project” has some missing values. 
Your task is to compute these missing values. Replace the missing values by running average.
You should use the “For” loop for this task.




```python
# loading the previously created project csv
df_project = pd.read_csv("project.csv")

# replacing the missing values in the Cost Column using running average
running_sum = 0
count = 0

for i in range(len(df_project)): #go thru each row
    if pd.isna(df_project.loc[i, "Cost"]): #using isna to detect missing values & loc for location in cells
          #if Nan detected appply the condition (running_sum/count) else return 0
        df_project.loc[i, "Cost"] = running_sum / count if count > 0 else 0 
       
    else:
        running_sum += df_project.loc[i, "Cost"] # to do the running total
        count += 1 # keep increasing the count by 1

# using astype() to avoid scientific notations (e.g. 1.002000e+06). astype() converts values into integers.
df_project["Cost"] = df_project["Cost"].astype(int) 

df_project.to_csv("project.csv", index=False)  # saving updated dataframe
print("Replaced missing values with running average:")
df_project
```

Task 3

Split the name column in the Employee dataframe into two new columns “First Name”,
and “Last Name” and remove the older “name” column

```python
df_employee = pd.read_csv("employee.csv") # loading employee csv

# spliting Name Column by " " to split the name into First Name & Last Name.
df_employee[['First Name', 'Last Name']] = df_employee['Name'].str.split(' ', expand=True)

# dropping Name column as asked
df_employee.drop(columns=['Name'], inplace=True)


df_employee.to_csv("employee.csv", index=False)  # now saving updated dataframe.
print(df_employee)
```

Task 4

Join all three dataframes in one single dataframe. Name it “Final”

```python
# loading all 3 csv files
df_employee = pd.read_csv("employee.csv")
df_seniority = pd.read_csv("seniority.csv")
df_project = pd.read_csv("project.csv")

# merging all datframes into Final by using "ID" as common entries within all dataframes
Final = df_project.merge(df_employee, on="ID").merge(df_seniority, on="ID")

Final.to_csv("Final.csv", index=False)  # saving Final dataframe.
print("Final :") #print statement for easy understanding
pd.DataFrame(Final)
```

Task 5

Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project
cost only to employees who have finished the projects.

```python
# calculating bonus based on the project cost only for "Finished" projects
Final['Bonus'] = Final.apply(lambda row: 0.05 * row['Cost'] if row['Status'] == 'Finished' else 0, axis=1)
Final.to_csv("Final.csv", index=False) #saving as csv
print(Final)

#created a column name Bonus.
#specifying the calculation for 5% bonus i.e. lambda row: 0.05*cost if status==finished.
#using apply() to operate lambda on every row.
#multiplying the row with 5% if condition is met otherwise display 0.
#axis=1 to apply these conditions vertically.
```

Task 6

Demote the designation level by 1, whose projects have status “fail”. Delete the
employees record whose designation level is above 4.

```python
projects_failed = Final[Final['Status'] == 'Fail'] #logic: detect the status==fail
Final.loc[projects_failed.index, 'Designation Level'] -= 1 # if detected, demote them by 1

# delete employees with Designation Level above 4
Final = Final[Final['Designation Level'] <= 4]

print(Final)
```

Task 7

Add “Mr.” and “Mrs.” to the first name column and drop the gender column.

```python
# addinng Mr/Mrs based on gender
#apply() for iterating lambda function on every row.
Final['First Name'] = Final.apply(lambda row: ("Mr. " if row['Gender'] == 'M' else "Mrs. ") + row['First Name'], axis=1)

# drop "Gender" column
Final.drop(columns=['Gender'], inplace=True)

Final.to_csv("Final.csv", index=False) #save Final as csv after changes
Final
```

Task 8

Promote designation level by 1 for the employees whose age is more than 29 years using
IF condition.

```python
for i in range(len(Final)): # using range with len for measuring "Final" length
    if Final.loc[i, 'Age']> 29: # search for the location of employee whose age is > 29
       Final.loc[i, 'Designation Level'] +=1 #if found, increase Designation Level by 1
pd.DataFrame(Final)
```

Task 9

Add the cost of all projects for each Employee and save it in new dataframe
“TotalProjCost” with three columns ID, First Name, and Total cost

```python
# calculating total project costs
# using groupby() to group according to ids, first name & total cost(using agg function sum() on Cost to get tc)
TotalProjCost = Final.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()

# rename() to change cost into total cost into TotalProjectCost DF
TotalProjCost.rename(columns={'Cost': 'Total Cost'}, inplace=True) 

# saving csv
TotalProjCost.to_csv("TotalProjCost.csv", index=False) 
print("TotalProjCost: ")
TotalProjCost
```

Print all the employee details whose city name contains the letter “o” in it
